距離$d$での電力密度は$\frac{1}{4\pi d^2} G_t P_t$なので、受信電力は$P_r = A_r \frac{1}{4\pi d^2} G_t P_t$になる。

実効面積と利得の関係には、アンテナの種類によらず、$A=\frac{\lambda^2}{4\pi} G$が成立することが知られている。
特に等方性アンテナの場合、$G=1, A=\frac{\lambda^2}{4\pi}$が成立する。
この値に対して定性的な理解をしたいところだが、どうもその背景を理解するのは難しそうである([フリスの伝達公式〜その周辺の不思議を探る〜 唐沢好男](http://www.radio3.ee.uec.ac.jp/ronbun/TR-YK-040_Friis-Formula.pdf))。

$L_p=(\frac{4\pi d}{\lambda})^2$は自由空間伝搬損と呼ぶ。



フリスの伝達公式が、sionnaで成立していることを、等方性アンテナを用いて確認する。

以下、receiverはダミーとして配置。

```python
scene = rt.Scene()

tx = rt.Transmitter(name="tx", position=[0, 0, 1])
scene.add(tx)
rx = rt.Receiver(name="rx", position=[0, 0, 10])
scene.add(rx)

scene.tx_array = rt.PlanarArray(num_rows=1, num_cols=1, pattern="iso", polarization="V")
scene.rx_array = rt.PlanarArray(num_rows=1, num_cols=1, pattern="iso", polarization="V")

p_solver = rt.PathSolver()
paths = p_solver(
    scene=scene,
    synthetic_array=False,
    seed=2913,
)

rm_solver = rt.RadioMapSolver()

distance = 1e3
cell_size = 1
rm = rm_solver(
    scene=scene,
    max_depth=5,
    center=[distance / 2, 0, 0],
    size=[distance, cell_size],
    orientation=[0, 0, 0],
    cell_size=[cell_size, cell_size],
    samples_per_tx=10**9,
)

fig = scene.preview(
    paths=paths,
    radio_map=rm,
)
```

```python
fig = scene.render(
    camera=rt.Camera(position=[-20,-20,20], look_at=[100,0,0]),
    paths=paths,
    radio_map=rm,
    num_samples=512,
    fov=90,
    rm_show_color_bar=True,
)
```
